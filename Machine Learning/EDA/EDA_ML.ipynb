```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
happiness_data = pd.read_csv('2015.csv')
happiness_data.head()
```

```python
# SHape of the  data (no of rows and columns)
happiness_data.shape
```

```python
# Statistics measurements
happiness_data.describe()
```

```python
# Checking missing values
missing_val = happiness_data.isnull().sum()
missing_val
```

```python
# checking duplicates
duplicate_values = happiness_data.duplicated()
duplicate_values.sum()
happiness_data[duplicate_values]
```

```python
# Checking Outliers
happiness_data.boxplot(column=['Happiness Score'])
plt.show()
```

```python
# Normalization 
from sklearn.preprocessing import StandardScaler
```
