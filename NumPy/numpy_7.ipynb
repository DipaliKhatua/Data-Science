### Find Trignometry sin(), cos(), tan() using Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.sin(180)
```

```python
np.sin(90)
```

```python
np.cos(180)
```

```python
np.tan(180)
```

```python
x_sign = np.arange(0,3*np.pi, 0.1)
print(x_sign)
```

```python
y_sign =- np.sin(x_sign)
print(y_sign)
```

```python
plt.plot(x_sign, y_sign)
plt.show()
```

```python
y_cos = np.cos(x_sign)
plt.plot(x_sign, y_cos)
plt.show()
```

```python
y_tan = np.tan(x_sign)
plt.plot(x_sign, y_tan)
plt.show()
```
