### Matplotlib Histogram

Matplotlib histogram is a representation of numeric data in the form of a rectangle bar. Each bar shows some data,  which belong to different categories. 

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

#### Generate Data

```python
ml_student_age = np.random.randint(18,45, (100))
py_student_age = np.random.randint(15, 40 , (100))
```

```python
print(ml_student_age)
```

```python
print(py_student_age)
```

#### Plotting Histogram Using Matplotlib

```python
plt.hist(ml_student_age)
plt.title("Machine Learning Student Age Histogram")
plt.xlabel("student age category")
plt.ylabel("No. of student age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth=0.8)
plt.title("Machine Learning Student Age Histogram")
plt.xlabel("student age category")
plt.ylabel("No. of student age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize = (16,9)) # size of histogram in 16:9 format
 
 
plt.hist(ml_student_age, bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = "m", label = "ML Student")
 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

#### Plotting two Histogram Using Matplotlib with parameters

```python
from matplotlib import style # for style 
style.use("ggplot") # return grid
plt.figure(figsize = (16,9)) 
 
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = ["m", "y"], label = ["ML Student", "Py Student"])
 
#plt.hist(py_students_age, bins, rwidth=0.8, histtype = "bar",
#         orientation='vertical', color = "y", label = "Py Student")
 
plt.title("ML & Py Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```
