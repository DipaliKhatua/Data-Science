### Matplotlib Scatter Plot

The plt.scatter() function help to plot two-variable datasets in point or a user-defined format. Here, we will be plotting google play store apps scatter plot.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_google_play_store_app = pd.read_csv('googleplaystore.csv', nrows=1000)
df_google_play_store_app.shape
```

```python
df_google_play_store_app.describe()
```

```python
df_google_play_store_app.columns
```

```python
x = df_google_play_store_app['Rating']
y = df_google_play_store_app['Reviews']
```

```python
plt.scatter(x , y)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x , y, c="r", marker="*", s = 200, alpha=0.5)
plt.title(" Google playstorew apps scatter plot")
plt.xlabel("rating")
plt.ylabel("Reviews")
plt.show()
```
